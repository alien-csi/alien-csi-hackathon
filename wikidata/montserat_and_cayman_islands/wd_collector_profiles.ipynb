# Fetch all statements from Wikidata
The snippet below fetches all statements linked to the above mentioned qids. All these triples are aggregated in a single data file (as RDF turtle triples).

````python
# Fetch all statements in wikidata and store as a turtle file
from rdflib import Graph
people = Graph()
for qid in qids:
  people.parse(qid+".ttl")
people.serialize(destination="/tmp/montseratandcaymanncollectors.ttl", format="turtle")
````

# Extract the Shape Expression
A shape expression is a formal description of a (subset of a) RDF file. 

````python
from shexer.shaper import Shaper
from shexer.consts import NT, TURTLE
import pprint

q = "select ?class where { ?item <http://www.wikidata.org/prop/direct/P31> ?class }"
target_classes = []
x = people.query(q)
for target_class in x:
  if str(target_class["class"]) not in target_classes:
    target_classes.append(str(target_class["class"]))

shex_target_file = "caymanandmontseratcollectorprofile.shex"

input_ttl_file = "/tmp/montseratandcaymanncollectors.ttl"

shaper = Shaper(target_classes=target_classes,
                graph_file_input=input_ttl_file,
                input_format=TURTLE,
                # namespaces_dict=namespaces_dict,  # Defaults: no prefixes
                instantiation_property="http://www.wikidata.org/prop/direct/P31",)  # Default rdf:type
            
shaper.shex_graph(output_file=shex_target_file)
````

Here is stated that in Wikidata on approx 80% of identified collectors their occupation is known.
